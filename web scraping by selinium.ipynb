```python
# Importing the webdriver module to interact with the browser
from selenium import webdriver 

# Automatically manages and installs the appropriate ChromeDriver
from webdriver_manager.chrome import ChromeDriverManager  

# Used to specify the path to the WebDriver executable (ChromeDriver in this case)
from selenium.webdriver.chrome.service import Service 

# Allows you to customize Chrome's behavior 
from selenium.webdriver.chrome.options import Options

# Importing By class from Selenium webdriver.common module
from selenium.webdriver.common.by import By 

# Setting up Chrome options (optional configuration for Chrome browser)
options = Options()

# Adding experimental option to keep the browser open after the script ends
options.add_experimental_option("detach", True)  

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
```

```python
# Get website
driver.get("https://www.flipkart.com/")
```

```python
# Find XPATH and then send keys on the search box
# search_box = driver.find_element(By.XPATH,"//input[@type='text']")
# search_box.send_keys("one plus headphones")
```

```python
# Find XPATH and then send keys on the search box
search_box = driver.find_element(By.CLASS_NAME,"Pke_EE")
search_box.send_keys("laptop")
```

```python
# Get attribute
first_option = driver.find_element(By.XPATH,'(//form[@class="_2rslOn header-form-search OpXDaO"]//a)[2]').click()
```

```python
# Display the current url 
print(driver.current_url)
```

```python
# Extract title data
product_title = []
title_path = driver.find_elements(By.XPATH,"//div[@class='KzDlHZ']")
for i in title_path:
    title = i.text
    product_title.append(title)

product_title = product_title[0:24]
print(product_title)
print(len(product_title))
    
```

```python
# Extract feature data
product_features = []
features_path = driver.find_elements(By.XPATH,'//div[@class="_6NESgJ"]//li[1]')
for i in features_path:
    features = i.text
    product_features.append(features)

features_title = product_features[0:24]
print(product_features)
print(len(product_features))
```

```python
# Extract RAM/ROM data
product_RAM = []
RAM_path = driver.find_elements(By.XPATH,'//div[@class="_6NESgJ"]//li[2]')
for i in RAM_path:
    RAM = i.text
    product_RAM.append(RAM)

RAM_title = product_RAM[0:24]
print(product_RAM)
print(len(product_RAM))
```

```python
# Extract operating_system data
product_operating_system = []
operating_system_path = driver.find_elements(By.XPATH,'//div[@class="_6NESgJ"]//li[3]')
for i in operating_system_path:
    operating_system = i.text
    product_operating_system.append(operating_system)

operating_system_title = product_operating_system[0:24]
print(product_operating_system)
print(len(product_operating_system))
```

```python
# Extract SSD data
product_SSD = []
SSD_path = driver.find_elements(By.XPATH,'//div[@class="_6NESgJ"]//li[4]')
for i in SSD_path:
    SSD = i.text
    product_SSD.append(SSD)

SSD_title = product_SSD[0:24]
print(product_SSD)
print(len(product_SSD))
```

```python
# Extract display data
product_display = []
display_path = driver.find_elements(By.XPATH,'//div[@class="_6NESgJ"]//li[4]')
for i in display_path:
    display = i.text
    product_display.append(display)

display_title = product_display[0:24]
print(product_display)
print(len(product_display))
```

```python
# Extract warranty data
product_warranty = []
warranty_path = driver.find_elements(By.XPATH,'//div[@class="_6NESgJ"]//li[4]')
for i in warranty_path:
    warranty = i.text
    product_warranty.append(warranty)

warranty_title = product_warranty[0:24]
print(product_warranty)
print(len(product_warranty))
```

```python
df=pd.DataFrame({
    "Product":product_title,
    "Features":product_features,
    "RAM/ROM":product_RAM,
    "Operating System":product_operating_system,
    "SSD":product_SSD,
    "Display":product_display,
    "Warranty":product_warranty
})
```
